```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
import numpy as np
import matplotlib.pyplot as plt

vendas = np.random.randint(1000, 3000, 50) #gera 50(qtd de numeros gerados) inteiros entre 1000 (min) e 3000 (max)
meses = np.arange(1, 51)

plt.plot(meses, vendas, 'pink')
plt.axis([0, 50, 0, max(vendas)+200])
plt.xlabel('meses')
plt.ylabel('vendas')
plt.show()
```

- Dispersão

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
import numpy as np
import matplotlib.pyplot as plt

vendas = np.random.randint(1000, 3000, 50) #gera 50(qtd de numeros gerados) inteiros entre 1000 (min) e 3000 (max)
meses = np.arange(1, 51)

plt.scatter(meses, vendas, color='pink')
plt.axis([0, 50, 0, max(vendas)+200])
plt.xlabel('meses')
plt.ylabel('vendas')
plt.show()
```

- Barras

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
import numpy as np
import matplotlib.pyplot as plt

vendas = np.random.randint(1000, 3000, 50) #gera 50(qtd de numeros gerados) inteiros entre 1000 (min) e 3000 (max)
meses = np.arange(1, 51)

plt.bar(meses, vendas, color='pink')
plt.axis([0, 50, 0, max(vendas)+200])
plt.xlabel('meses')
plt.ylabel('vendas')
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
import numpy as np
import matplotlib.pyplot as plt

vendas = np.random.randint(1000, 3000, 50) #gera 50(qtd de numeros gerados) inteiros entre 1000 (min) e 3000 (max)
meses = np.arange(1, 51)

#plt.bar(meses, vendas)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(meses, vendas)
plt.subplot(1, 3, 2)
plt.scatter(meses, vendas)
plt.subplot(1, 3, 3)
plt.bar(meses, vendas)
plt.show()
```
