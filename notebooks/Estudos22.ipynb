# Como instalar módulos novos
- usamos o 'pip' para isso. É um gerenciador de pacotes que nos ajuda a instalar/desinstalar/atualizar módulos
    - passo 1: abrir prompt anaconda
    - passo 2: instalar com o pip os pacotes que quiser
    - passo 3: reiniciar o jupyter (se necessario) ctrl+c no promp e abre de novo

```python
import keyboard

keyboard.press_and_release('ctrl+t')
keyboard.write('https://devvbie.github.io/curriculum/')
keyboard.press_and_release('enter')
```

```python
pip install keyboard
pip uninstall keyboard
```
